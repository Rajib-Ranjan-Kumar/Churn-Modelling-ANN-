```python
import pandas as pd
import numpy as np
from tensorflow.keras.models import load_model

```

```python
model =load_model("model.h5")
```

```python
import pickle
with open("C:\ds\churn_modelling\label_encoder_gender.pkl",'rb') as file:
    label_encoder_gener=pickle.load(file)
with open("C:\ds\churn_modelling\onehot_encoder_geo.pkl",'rb') as file:
    onehot_encoder_geo=pickle.load(file)
with open("C:\ds\churn_modelling\scaler.pkl",'rb') as file:
    scaler=pickle.load(file)
```

```python
input_data = {
'CreditScore': 0,
'Geography': 'France',
'Gender': 'Male',
'Age': 40,
'Tenure': 3,
'Balance': 0,
'NumOfProducts': 0,
'HasCrCard': 1,
'IsActiveMember': 1,
'EstimatedSalary': 5
}
```

```python
df=pd.DataFrame([input_data],columns=input_data.keys())
```

```python
df.head()
```

```python
df['Gender']=label_encoder_gener.transform(df['Gender'])
```

```python
df.head()
```

```python
geo_df = onehot_encoder_geo.transform(df[['Geography']]).toarray()
geo_df=pd.DataFrame(geo_df,columns=onehot_encoder_geo.get_feature_names_out())
```

```python
geo_df.head()
```

```python
df=pd.concat([df,geo_df],axis=1)
df.drop(columns=['Geography'],axis=1,inplace=True)
```

```python
df.head()
```

```python
df=pd.DataFrame(scaler.transform(df),columns=scaler.get_feature_names_out())
```

```python
df.head()
```

```python
y_predict=model.predict(df)
```

```python
y_predict[0][0]
```
